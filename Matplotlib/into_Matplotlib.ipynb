### Labeling graph

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years,runs)
plt.title("Sachin's Runs vs Year")
plt.xlabel("Year")
plt.ylabel("Runs")
```

### Multiple lines in one graph

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years,kohli, label="Virat Kohli")
plt.plot(years , sehwag , label="Sehwag")
plt.xlabel("Year")
plt.ylabel("Run scored")
plt.title("Performance Comparison")
plt.legend()
```

### Using Format Strings

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
plt.plot(years,kohli,'ro--',label="Virat Kohli")
plt.plot(years,sehwag,'b^:',label="Virendra Sehwag")
plt.xlabel("Year")
plt.ylabel("Run scored")
plt.title("Performance Comparison")
plt.legend()
```

```python
# more readable syantax
plt.plot(years, kohli, color='orange', linestyle='--',marker="v", label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.',marker="<", label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar",linewidth=4)
plt.legend()
```

```python
plt.grid(True)
plt.plot(years, kohli, color='orange', linestyle='--',marker="v", label="Kohli")
plt.plot(years, sehwag, color='green', linestyle='-.',marker="<", label="Sehwag")
plt.plot(years, runs, color='blue', label="Tendulkar",linewidth=4)
plt.legend()
```

```python
#XKCD comic style
with plt.xkcd():
    plt.plot(years, kohli, label="Kohli")
    plt.plot(years, sehwag, label="Sehwag")
    plt.title("Epic Battle of the Batsmen")
    plt.legend()
    plt.show()
```

```python
# big data creats confusion
import numpy as np

for i in range(50):
    plt.plot(np.random.rand(100), linewidth=1)

plt.title("Too Much Data Can Be Confusing!")
plt.grid(True)
plt.tight_layout()
plt.show()
```
