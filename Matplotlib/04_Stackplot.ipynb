```python
import matplotlib.pyplot as plt 
```

```python
days = [1, 2, 3, 4, 5, 6, 7]  # Days of the week
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]
labels=['studying','playing','watching_tv','sleeping ']
colours=['red','blue','yellow','green']
```

```python
plt.figure(figsize=(10,6))
plt.stackplot(days, studying, playing, watching_tv, sleeping,labels=labels,colors=colours,alpha=0.6)
plt.legend(loc='upper right')
plt.xlabel("Days")
plt.ylabel("Hours")
plt.title("Hours spend by student")
plt.grid(True)
```
