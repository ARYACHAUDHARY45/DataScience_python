```python
## Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.
   #Then create a horizontal bar chart showing total runs scored in their debut 5 years.
```

```python
import matplotlib.pyplot as plt
import numpy as np
Sachin=[215,612,495,1123,959]
Kohli=[159,325,1021,1644,2186]
Sehwag=[1,19,674,1767,1393]
years=np.array([1,2,3,4,5]) #debut year of cricketer
width=0.25
```

```python
plt.bar(years-0.25,Sachin,width=width,label="Sachin")
plt.bar(years,Kohli,width=width,label="Kohli")
plt.bar(years+0.25,Sehwag,width=width,label="Sehwag")
plt.xlabel("Debut Year")
plt.ylabel("Runs scored")
plt.legend()
plt.title("Run scored in first 5 years")
```
