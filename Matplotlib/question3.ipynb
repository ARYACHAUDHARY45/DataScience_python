```python
#Create a pie chart showing the market share of mobile OS (Android, iOS, others).
#Then recreate the same using a horizontal bar chart and observe which is easier to understand
```

```python
import matplotlib.pyplot as plt
OS = ['Android',' IOS ','Samsung']
market_share = [68.24,31.48, 0.28]
colors=['#3DDC84','#007AFF','#1428A0']
explode=[0.1,0,0]
plt.pie(market_share, labels =  OS,colors =colors,wedgeprops={'edgecolor' : 'Black'},explode=explode,autopct='%1.1f%%' )
plt.title('Market share of different OS in the World')
```

```python
#Horizontal bar chat
OS = ['Android',' IOS ','Samsung']
market_share = [68.24,31.48,0.28]
plt.barh(OS,market_share)
```
