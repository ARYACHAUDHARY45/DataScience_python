##Numpy vs python list


```python
#using python list
import time
size=1_000_000_00
l1=list(range(size))
l2=list(range(size))
start=time.time()
result=[x+y for x,y in zip(l1,l2)]
end=time.time()
print("Time =",end-start)
```

```python
#using numpy 
import numpy as np
import time
size=1_000_000_00
l1=list(range(size))
l2=list(range(size))
array1=np.array(l1)
array2=np.array(l2)
start=time.time()
result=array1+array2
end=time.time()
print("Time =",end-start)
```

```python
#squares in lists and numpy
import numpy as np
import time
l1=list(range(100))
start=time.time()
result=[x**2 for x in l1]
end = time.time()
print("time taken in list is =",end-start)

array1=np.array(l1)
start=time.time()
result=array1**2
end=time.time()
print("time taken in Numpy is =",end-start)
```
