### Q:1 Create a 3×3 array filled with random numbers and print its shape.

```python
import numpy as np
arr=np.random.rand(3,3)
print(arr)
print("Shape Of Array Is = ",arr.shape)
```

### Q:2 Convert an array of floats [1.1, 2.2, 3.3] into integers.

```python
floatarray=np.array([1.1,2.2,3.3],dtype=int)#Dtype used to convert array when it created
print(floatarray)
arr=np.array([1.1,2.2,3.3])
int_array=arr.astype(int)
print(int_array)
```

### Q:3 Use fancy indexing to extract even numbers from [1, 2, 3, 4, 5, 6].

```python
arr=np.array([1,2,3,4,5,6])
print(arr[[1,3,5]])
```

### Q:4 Reshape a 1D array of size 9 into a 3×3 matrix.

```python
arr=np.arange(1,10)#1D array of 9 elements
matrix=arr.reshape(3,3)
print(matrix)
```

### Q:5 Use boolean masking to filter numbers greater than 50 in an array.

```python
arr=np.arange(40,60,2)
print(arr[arr>50])
```
