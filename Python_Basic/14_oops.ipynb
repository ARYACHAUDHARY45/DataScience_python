## OOP in Python.

```python
class employee:
    company="jio"
    def __init__(self,name,age):
        self.name= name
        self.age=age
    @staticmethod  #This is static method 
    def printtime():
        print("Time is money...")
    @classmethod #This is for class atribute.
    def classdetail(cls):
        print(f"Company is {cls.company}....")
    def printdetails(self):
        print(f"Name is {self.name} and age is {self.age} years.")
```

```python
e= employee("Arya" , 20)
```

```python
e.printdetails()
```

```python
e.printtime()
```

```python
e.classdetail()
```
