```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
    if(fruit=="banana"):
        break
```

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    if(fruit=="banana"):
        continue
    print(fruit)
```
