## Match case Conditional statements

```python
# used for pattern matching .
```

```python
code = 500
```

```python
match code :
    case 200:
        print("Response ok")
    case 400:
        print("Bad response")
    case 500:
        print("khel khatam beta...")
    case _:
        print("nothing")
     
```
