## Lambda function in python

```python
cube = lambda x : x*x*x
```

```python
cube(3)
```

```python
s= lambda a,b : a+b
```

```python
s(1,9)
```

```python
numbers=[1,2,3,4]
```

```python
square = list(map(lambda x:x*x , numbers))  #map method to iterate list..
```

```python
square
```

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4, 6]
```

```python
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24
```

```python
students = [("Alice", 85), ("Bob", 78), ("Charlie", 92)]
students.sort(key=lambda student: student[1])  # Sort by score
print(students)  # Output: [('Bob', 78), ('Alice', 85), ('Charlie', 92)] #for sorting
```
