## list comprehensions

```python
Table=[]
```

```python
for i in range(1,11):
    print(f" 2 x {i} = {2*i}")
    Table.append(2*i)
```

```python
Table
```

### now using list comprehensions

```python
t=[2*i for i in range(1,11)]
```

```python
t
```

```python
t=[2*i for i in range(1,11) if i%2==0]
```

```python
t
```
