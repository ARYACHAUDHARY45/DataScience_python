## Data Selection

```python
import pandas as pd
```

```python
df=pd.read_excel("data.xlsx")
```

```python
df
```

```python
#Selecting single Columns
df["Actor"]
```

```python
#Selecting Multiple Columns
df[['Actor','IMDb']]
```

```python
#Selecting rows by label
df.loc[0]
```

```python
#Selecting rows by position
df.iloc[0]
```

```python
#specific column
df.loc[7,'Actor']
```

```python
#slcing 
df.loc[0:2,['Actor','Film','Genre']] #2 is inclusive
```

```python
df.iloc[0:2,0:2] #2 is exclusive
```

```python
#for fast access to a perticular element
df.at[7,'Actor']
```

```python
df.iat[7,0]
```

## Data Filtering

```python
df
```

```python
#for simple condition
df[df['IMDb']>6]
```

```python
#for multiple condition
df[(df['IMDb']>6) & (df['Year']>2018)]
```

```python
df[(df['IMDb']>6) | (df['Year']>2018)]
```

```python
# Use of .query() 
df.query("Actor=='Ranbir Kapoor'")
```

```python
df.query("Year > 2018 and IMDb > 7") #multiple condition
```

```python
    #Use backticks for column names with spaces or special characters
```

```python
#use @ to reference Python variables
y=2018
df.query("Year > @y and IMDb > 7")
```

```python
#Chained comparisons
df.query("2018 < Year < 2020")
```

```python
#.query() returns a copy, not a view
filter=df.query("2018 < Year < 2020")
print(filter)
```
