## Core Data-structure

### Series DS

```python
import pandas as pd
pd.Series([10,20,30,40,50])
#0,1,2,3,4 are auto mated index
```

```python
#we can give custom index
s=pd.Series([10,20,30,40,50],index=["Arya","Priyam","Jinil","Tushar","Bhavy"])
print(s)
```

```python
type(s)
```

### DataFrame — 2D Labeled Table

```python
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "city": ["Delhi", "Mumbai", "Bangalore"]
}

df = pd.DataFrame(data)
print(df)
```
