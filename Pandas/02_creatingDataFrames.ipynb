## Creating dataframes

```python
import pandas as pd
```

```python
#from python list
data = [
    ["Alice", 25],
    ["Bob", 30],
    ["Charlie", 35]
]
df=pd.DataFrame(data, columns=["Name","Age"])
print(df)
```

```python
#from Dictionary of lists
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35]
}
df=pd.DataFrame(data)
print(df)
```

```python
#from Numpy array
import numpy as np
arr=np.array([[1,2,3],[4,5,6]])
df=pd.DataFrame(arr,columns=["A","B","C"])
print(df)
```

```python
#from xlsx file
df=pd.read_excel("data.xlsx")
print(df)
```

```python
#from Csv file
df=pd.read_csv("data.csv")
print(df)
```
