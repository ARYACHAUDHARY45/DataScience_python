### Sorting and Ranking

```python
import pandas as pd
df=pd.read_excel("data.xlsx")
```

```python
df
```

```python
df2 = df.sort_values("Year") # Ascending sort
```

```python
df2 = df.sort_values("Year",ascending=False) # decending sort
```

```python
df2
```

```python
df2 = df.sort_values(["Year","IMDb"]) # Sort by multiple columns
```

```python
df2
```

```python
df2.reset_index(drop=True,inplace=True)
```

```python
df2
```

```python
df2["Rank"]=df2["IMDb"].rank(method="dense",ascending=False)
```

```python
df2
```
