## Data Cleaning & Preprocessing

```python
import pandas as pd
```

```python
df=pd.read_excel("data_cleaning_sample.xlsx")
```

```python
df
```

```python
df.isnull()              # True for NaNs
```

```python
df.isnull().sum()        # Count missing per column
```

```python
df.dropna() # Drop rows with *any* missing values
```

```python
df.dropna(axis=1) # Drop columns with missing values
```

```python
df.fillna(0) ## Replace NaN with 0
```

```python
df["Age"].fillna(df['Age'].mean())
```

```python
df.ffill() #forward fill
```

```python
df.bfill() #backwardfill
```

```python
#Detecting and removing duplicates
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
# for prticular columns , we can also check.
df.duplicated(subset=["Name","Age"])
```

```python
df.drop_duplicates(subset=["Name","Age"])
```

```python
#String Operations with .str
df["Name"].str.lower()
```

```python
df['City'].str.contains("delhi",case=False)  #case=false means case insensitive
```

```python
df["Email"].str.split("@")
```

```python
type(df["Email"].str.split("@")[0]) # stores in list
```

```python
#Type conversion usinf .astype()
df2=df.dropna().copy()
```

```python
df2
```

```python
df2["Age"]=df2["Age"].astype(int)
```

```python
df2
```

```python
# Applying functions
df2["Age Group"] = df['Age'].apply(lambda x: "Adult" if x>27 else "Minor")
```

```python
df2
```

```python
gender_map = {"M": "Male", "F": "Female"}
df2["Gender"]=df2["Gender"].map(gender_map)
```

```python
df2
```

```python
df2["City"]=df2["City"].replace({"Delhi": "New Delhi", "Mumbai": " New Mumbai"})
```
